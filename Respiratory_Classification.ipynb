##Import Library

```python
# อ่านข้อมูลเสียง
import librosa, librosa.display

# จัดการข้อมูลและการทำกราฟ
import numpy as np
import matplotlib.pyplot as plt

# จัดการไฟล์เสียง
from IPython.display import Audio
from scipy.io import wavfile

# แสดง progress การทำงาน
from tqdm import tqdm

import warnings
warnings.filterwarnings('ignore')
```

##ทดสอบการอ่านไฟล์เสียง

ใช้ librosa อ่านไฟล์เสียง Disease (100).wav และแสดงผลเป็นกราฟแบบ spectrogram (กราฟแสดงข้อมูลเสียงแบบ spectrum)

```python
f , sr = librosa.load(Disease_data + '/Disease_D-1.wav') # อ่านไฟล์เสียง
st = librosa.stft(f) # แปลงข้อมูลเสียง ด้วยฟังก์ชั่น stft (Short-time Fourier transform)
spectogram = np.abs(st)
plt.imshow(librosa.amplitude_to_db(spectogram))
Audio(Disease_data + '/Disease_D-1.wav')
```

##เขียนฟังก์ชั่นแปลง stft

ดูมิติของเสียง ซึ่งไฟล์เสียง Disease (100).wav มีขนาด 1025 แถว 493 คอลัมน์


```python
print(spectogram.shape)
```

```python
def spect(wave):
    f, sr = librosa.load(wave , duration = 0.9)
    st = librosa.stft(f)
    spectogram = np.abs(st)
    return spectogram
```

##อ่านข้อมูล ไฟล์เสียงทีละไฟล์

```python
# ประกาศตัวแปร เอาไว้ training
x = list()
y = list()

# อ่านชื่อไฟล์ที่อยู่โฟลเดอร์ของการเป็นโรคและปกติ
ds_files = os.listdir(Disease_data)
nm_files = os.listdir(Normal_data)

# อ่านข้อมูลทีละไฟล์ 
# ทำ label แบบ one-hot encoder 
# ให้เสียงหายใจของคนเป็นโรคเป็น [0,1] และ เสียงหายใจของคนปกติเป็น [1,0]
for f in tqdm(ds_files):
    file1 = Disease_data + f
    data = spect(file1) #อ่านข้อมูลเสียงทีละไฟล์
    x.append(data)  # อัดเข้าตัวแปร x
    y.append([0 , 1]) # กำหนด label 0,1 คือ โรค เข้าตัวแปร y

for f in tqdm(nm_files):
    file1 = Normal_data + f
    data = spect(file1) #อ่านข้อมูลเสียงทีละไฟล์
    x.append(data)  # อัดเข้าตัวแปร x
    y.append([1 , 0]) #กำหนด label 1,0 คือ ปกติ เข้าตัวแปร y
```

##เตรียมข้อมูลการ train

Import Library

```python
from sklearn.model_selection import train_test_split
```

##แบ่งชุดข้อมูลเป็น train 80% และ test 20%

```python
X = np.array(x)
y = np.array(y)

X = X.reshape(X.shape[0] , X.shape[1] , X.shape[2] , 1)
xtrain , xtest , ytrain , ytest = train_test_split(X , y , train_size = 0.8 , random_state = 16 )
```

##ดู shape ของข้อมูลที่ได้มา

```python
xtrain.shape , xtest.shape , ytrain.shape , ytest.shape
```

------- traning data 268 ไฟล์ และ testing data 68 ไฟล์ โดยแต่ละไฟล์มีมิติคือ (1025 , 39 , 1) ส่วนมิติของตัว label ก็จะเป็น (268 , 2) และ (68 , 2) ซึ่งก็คือจำนวนไฟล์ กับ label ที่เรา encode ไว้สองหลัก (0 , 1) นั่นเอง -------

#Training Model

##Import Library

```python
import tensorflow

import keras
from keras.models import Sequential
from keras.layers import Dense , Dropout , Activation , Flatten
from keras.layers import Conv2D , MaxPooling2D , Dropout
```

```python
model = keras.Sequential([
    keras.layers.Conv2D(128 , 
                        (3 , 3) , 
                        activation = 'relu' , 
                        input_shape = xtrain.shape[ 1 : ]) ,
    keras.layers.MaxPooling2D(pool_size = (2 , 2)) ,
    keras.layers.Conv2D(128 , 
                        (3 , 3) , 
                        activation = 'relu') ,
    keras.layers.MaxPooling2D(pool_size=(2 , 2 )) ,
    keras.layers.Dense(16) ,
    keras.layers.Dense(8) ,
    keras.layers.Flatten() ,
    keras.layers.Dense(2 , activation = 'softmax')
])

model.compile(optimizer = keras.optimizers.RMSprop(lr = 0.0001) , 
              loss = 'categorical_crossentropy' , 
              metrics= ['accuracy'])

batch_size = 6
epochs = 2000
```

##Training

```python
history = model.fit(xtrain , 
                    ytrain , 
                    epochs = epochs ,
                    validation_data = (xtest , ytest))
```

#Plot Graph

##Import Library

```python
import plotly.graph_objects as go
```

##Plot Loss

```python
h1 = go.Scatter(y = history.history['loss'] , 
                    mode = "lines" , line = dict(
                    width = 2 ,
                    color = 'blue') ,
                    name = "loss"
                   )

h2 = go.Scatter(y = history.history['val_loss'] , 
                    mode = "lines" , line = dict(
                    width = 2 ,
                    color = 'red') ,
                    name = "val_loss"
                   )

data = [h1 , h2]
layout1 = go.Layout(title = 'Loss' ,
                   xaxis = dict(title = 'epochs') ,
                   yaxis = dict(title = ''))

fig1 = go.Figure(data = data , layout = layout1)
fig1.show()
```

##Plot Accuracy Graph

```python
h1 = go.Scatter(y=history.history['accuracy'] , 
                    mode = "lines" , line = dict(
                    width = 2 ,
                    color = 'blue') ,
                    name = "acc"
                   )

h2 = go.Scatter(y=history.history['val_accuracy'] , 
                    mode = "lines" , line = dict(
                    width = 2 ,
                    color = 'red') ,
                    name = "val_acc"
                   )

data = [h1 , h2]
layout1 = go.Layout(title = 'Accuracy' ,
                   xaxis = dict(title = 'epochs') ,
                   yaxis = dict(title = ''))

fig2 = go.Figure(data = data, layout = layout1)
fig2.show()
```

```python
score = model.evaluate(xtrain, ytrain, verbose = 0)
print("Training Accuracy : " , score[1])
```
